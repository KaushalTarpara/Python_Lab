<h3>1.Create a file results.csv with fields: rollno, name, cmarks, cppmarks, pymarks. Insert at least 10 rows into that file.
Load csv data into a DataFrame.
Compute total marks.
Compute percentage.
Compute grade.
Plot line chart: name -> percentage. Show all chart attributes.
Plot bar chart: rollno:cmarks, cppmarks, pymarks -> percentage. Show all chart attributes.
Plot pie chart: Subject viz. count-of-students-who-scored-more-than-80-percentage.
Save updated DataFrame into a file: updated-result.csv.

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

# Create a CSV file with student data (results.csv)
data = [
    ["rollno", "name", "cmarks", "cppmarks", "pymarks"],
    [1, "Kaushal", 90, 85, 78],
    [2, "Anuj", 75, 88, 92],
    [3, "Raj", 82, 79, 85],
    [4, "Jenish", 88, 90, 76],
    [5, "Savan", 78, 82, 95],
    [6, "Aman", 91, 76, 88],
    [7, "Karan", 85, 92, 89],
    [8, "Jay", 92, 85, 94],
    [9, "Viru", 76, 78, 92],
    [10, "Shiv", 89, 91, 87]
]

with open('results.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

# Load CSV data into a DataFrame
df = pd.read_csv('results.csv')

# Calculate total marks and percentage
df['total'] = df['cmarks'] + df['cppmarks'] + df['pymarks']
df['percentage'] = (df['total'] / 300) * 100

# Define a function to compute grades
def compute_grade(percentage):
    if percentage >= 90:
        return 'A+'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B'
    elif percentage >= 60:
        return 'C'
    else:
        return 'F'

# Apply the function to compute grades
df['grade'] = df['percentage'].apply(compute_grade)

# Save the updated DataFrame to a new CSV file
df.to_csv('updated-result.csv', index=False)

# Line chart: name -> percentage
plt.figure(figsize=(10, 5))
plt.plot(df['name'], df['percentage'], marker='o', linestyle='-', color='b')
plt.title('Student Performance')
plt.xlabel('Name')
plt.ylabel('Percentage')
plt.xticks(rotation=45, fontsize=10)
plt.grid(True)
plt.tight_layout()

# Bar chart: rollno:cmarks, cppmarks, pymarks -> percentage
plt.figure(figsize=(10, 5))
plt.bar(df['rollno'], df['cmarks'], width=0.2, label='C Marks', align='center')
plt.bar(df['rollno'], df['cppmarks'], width=0.3, label='CPP Marks', align='center')
plt.bar(df['rollno'], df['pymarks'], width=0.2, label='PY Marks', align='edge')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.title('Subject-wise Marks')
plt.legend()
plt.xticks(df['rollno'])
plt.grid(axis='y')
plt.tight_layout()

# Count students who scored more than 80% in any subject
high_scorers = df[(df['cmarks'] > 80) | (df['cppmarks'] > 80) | (df['pymarks'] > 80)]
subject_count = high_scorers.shape[0]

# Pie chart: Subject viz. count-of-students-who-scored-more-than-80-percentage
labels = ['High Scorers', 'Others']
sizes = [subject_count, len(df) - subject_count]
colors = ['skyblue', 'lightcoral']
explode = (0.1, 0)

plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Students Scoring More Than 80% in Any Subject')
plt.axis('equal')

# Show the plots
plt.show()
```

<h3>2. Create a 3X4 array of strings. Store data into it. Print it on screen.
Create a 2X5 array of zeros.
Create a 2X5X3 array of ones.
Using suitable examples, demonstrate reshaping of arrays which were been created in 2-A, 2-B and 2-C.

```python
import numpy as np

# 2-A: Create a 3x4 array of strings and print it
array_2a = np.array([["A", "B", "C", "D"],
                     ["E", "F", "G", "H"],
                     ["I", "J", "K", "L"]])
print("2-A: 3x4 array of strings:")
print(array_2a)

# 2-B: Create a 2x5 array of zeros
array_2b = np.zeros((2, 5))
print("\n2-B: 2x5 array of zeros:")
print(array_2b)

# 2-C: Create a 2x5x3 array of ones
array_2c = np.ones((2, 5, 3))
print("\n2-C: 2x5x3 array of ones:")
print(array_2c)

# Reshaping examples:
# Reshape the 3x4 array (2-A) into a 4x3 array
reshaped_2a = array_2a.reshape(4, 3)
print("\nReshaped 2-A (3x4 to 4x3):")
print(reshaped_2a)

# Reshape the 2x5 array of zeros (2-B) into a 5x2 array
reshaped_2b = array_2b.reshape(5, 2)
print("\nReshaped 2-B (2x5 to 5x2):")
print(reshaped_2b)

# Reshape the 2x5x3 array of ones (2-C) into a 3x2x5 array
reshaped_2c = array_2c.reshape(3, 2, 5)
print("\nReshaped 2-C (2x5x3 to 3x2x5):")
print(reshaped_2c)
```

<h3>3. Explain purpose and demostrate usage of following functions:
dot()
arange()
hstack()
vstack()
linspace()
logspace()
asarray()
matmul()

```python
import numpy as np

# 1. dot()
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result_dot = np.dot(A, B)
print("1. dot():")
print(result_dot)
print()

# 2. arange()
arr_arange = np.arange(0, 10, 2)
print("2. arange():")
print(arr_arange)
print()

# 3. hstack()
A_hstack = np.array([1, 2, 3])
B_hstack = np.array([4, 5, 6])
result_hstack = np.hstack((A_hstack, B_hstack))
print("3. hstack():")
print(result_hstack)
print()

# 4. vstack()
A_vstack = np.array([1, 2, 3])
B_vstack = np.array([4, 5, 6])
result_vstack = np.vstack((A_vstack, B_vstack))
print("4. vstack():")
print(result_vstack)
print()

# 5. linspace()
arr_linspace = np.linspace(0, 1, 5)
print("5. linspace():")
print(arr_linspace)
print()

# 6. logspace()
arr_logspace = np.logspace(0, 2, 5)
print("6. logspace():")
print(arr_logspace)
print()

# 7. asarray()
list_data = [1, 2, 3]
arr_asarray = np.asarray(list_data)
print("7. asarray():")
print(arr_asarray)
print()

# 8. matmul()
A_matmul = np.array([[1, 2], [3, 4]])
B_matmul = np.array([[5, 6], [7, 8]])
result_matmul = np.matmul(A_matmul, B_matmul)
print("8. matmul():")
print(result_matmul)
print()
```
